#Praktikum 1

```python
def bubble_sort(data):
    n = len(data)
    for i in range(n - 1): # Corrected outer loop for passes
        for j in range(0, n - i - 1): # Inner loop for comparisons
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data # Correctly indented

angka =[64, 34, 25, 12, 22, 11, 90]
print ("Data sebelum diurutkan:", angka)
print ("Data setelah Bubble Sort:", bubble_sort(angka))
```

#Praktikum 2

```python
def selection_sort(data):
  n = len(data)
  for i in range(n):
    min_index = i
    for j in range (i + 1, n):
      if data[j] < data[min_index]:
        min_index = j
    data[i], data[min_index] = data[min_index], data[i]
  return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

#Praktikum 3

```python
def insertion_sort(data):
    n = len(data)
    for i in range(1, n):
      key = data[i]
      j = i - 1
      while j >= 0 and data[j] > key:
        data[j + 1] = data[j]
        j -= 1
      data[j + 1] = key
    return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

#Praktikum 6

```python
def bubble_sort_key(data, key_index=0, reverse=False):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if reverse:
                if data[j][key_index] < data[j + 1] [key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
            else:
                if data[j][key_index] > data[j + 1] [key_index]:
                    data[j], data[j + 1] = data[j + 1], data[j]
    return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]

print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(),0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

#Eksperimen Bubble Short

```python
def bubble_sort_dec(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n - i - 1):
            if data[j] < data[j + 1]:  # Untuk urut menurun
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

angka = [5, 2, 9, 1, 7, 3]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort descending:", bubble_sort_dec(angka))
```

#Eksperimen Selection Short

```python
def selection_sort_max_to_end(data):
    n = len(data)
    for i in range(n - 1):
        min_index = i
        # Cari elemen terbesar
        for j in range(i + 1, n):
            if data[j] < data[min_index]:
                min_index = j
        # Tukar posisi
        data[i], data[min_index] = data[min_index], data[i]
    return data

angka = [5, 2, 9, 1, 7, 3]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort Max :", selection_sort_max_to_end(angka))
```

# Eksperimen Insertion Sort

```python
def insertion_sort_nama(data):
    n = len(data)
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

mahasiswa = ["Budi", "Andi", "Citra", "Dewi", "Eka"]
print("Data sebelum diurutkan:", mahasiswa )
print("Data setelah Insertion Sort", insertion_sort_nama(mahasiswa))
```

# Analisis Kompleksitas

```python
import random

# Bubble Sort Desc + hitung perbandingan
def bubble_sort_desc(data):
    compares = 0
    n = len(data)

    for i in range(n):
        for j in range(0, n - i - 1):
            compares += 1
            if data[j] < data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]

    return data, compares


# Selection Sort Max + hitung perbandingan
def selection_sort_max(data):
    compares = 0
    n = len(data)

    for i in range(n):
        max_idx = i
        for j in range(i + 1, n):
            compares += 1
            if data[j] > data[max_idx]:
                max_idx = j
        data[i], data[max_idx] = data[max_idx], data[i]

    return data, compares


# Insertion Sort + hitung perbandingan
def insertion_sort_string(data):
    compares = 0

    for i in range(1, len(data)):
        key = data[i]
        j = i - 1

        while j >= 0:
            compares += 1
            if data[j] > key:
                data[j + 1] = data[j]
                j -= 1
            else:
                break

        data[j + 1] = key

    return data, compares


# Fungsi hitung perbandingan
def hitung_perbandingan(func, size):
    data = [random.randint(1, 1000) for _ in range(size)]
    _, compares = func(data.copy())
    return compares

ukuran = [10, 50, 100]

print("=== Perbandingan Bubble Sort ===")
for u in ukuran:
    print(u, "→", hitung_perbandingan(bubble_sort_desc, u))

print("\n=== Perbandingan Selection Sort ===")
for u in ukuran:
    print(u, "→", hitung_perbandingan(selection_sort_max, u))

print("\n=== Perbandingan Insertion Sort ===")
for u in ukuran:
    print(u, "→", hitung_perbandingan(insertion_sort_string, u))
```

```python
# Program Manajemen Barang Sederhana

def tampilkan_barang(data):
    print("\nDaftar Barang:")
    for nama, harga in data:
        print(f"{nama} - Rp{harga}")

def urutkan_barang(data, berdasarkan, urutan):
    # berdasarkan: "nama" atau "harga"
    # urutan: "asc" atau "desc"

    index = 0 if berdasarkan == "nama" else 1
    reverse = True if urutan == "desc" else False

    return sorted(data, key=lambda x: x[index], reverse=reverse)


# ===== Program Utama =====
barang = []

jumlah = int(input("Masukkan jumlah barang: "))

for i in range(jumlah):
    nama = input(f"Nama barang ke-{i+1}: ")
    harga = int(input(f"Harga barang ke-{i+1}: "))
    barang.append((nama, harga))

while True:
    print("\nMenu:")
    print("1. Urutkan berdasarkan nama (ASC)")
    print("2. Urutkan berdasarkan nama (DESC)")
    print("3. Urutkan berdasarkan harga (ASC)")
    print("4. Urutkan berdasarkan harga (DESC)")
    print("5. Keluar")

    pilihan = input("Pilih menu (1-5): ")

    if pilihan == "1":
        barang = urutkan_barang(barang, "nama", "asc")
        tampilkan_barang(barang)

    elif pilihan == "2":
        barang = urutkan_barang(barang, "nama", "desc")
        tampilkan_barang(barang)

    elif pilihan == "3":
        barang = urutkan_barang(barang, "harga", "asc")
        tampilkan_barang(barang)

    elif pilihan == "4":
        barang = urutkan_barang(barang, "harga", "desc")
        tampilkan_barang(barang)

    elif pilihan == "5":
        print("Program selesai.")
        break

    else:
        print("Pilihan tidak valid!")
```
